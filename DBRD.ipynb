```python
from sentence_transformers.readers import InputExample 
from torch.utils.data import DataLoader
from sentence_transformers import SentenceTransformer,  SentencesDataset, LoggingHandler, losses#train 
```

```python
model = SentenceTransformer('bert-base-uncased')
train_examples = [InputExample(texts=list(x))for x in train_data_array]
train_dataset = SentencesDataset(train_examples, model)
train_dataloader = DataLoader(train_dataset, shuffle=True, batch_size=2)
train_loss = losses.TripletLoss(model=model)
```

```python
#Tune the model
model.fit(train_objectives=[(train_dataloader, train_loss)], epochs=100, warmup_steps=100)
```
